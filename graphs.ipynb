```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import matplotlib
import matplotlib.font_manager as fm

# Data
modulos = [
    'lib/stepCalc',
    'components/Algorithms', 
    'components/Layout',
    'components/Tabs',
    'components/Structures',
    'components/Cards',
    'components/Info',
    'routes',
    'Outros (Icons, lib, etc.)'
]

linhas = [4346, 1780, 1968, 1521, 1573, 1382, 891, 301, 2713]
arquivos = [10, 12, 16, 9, 5, 7, 11, 5, 52]

# Calculate proportional areas
total_linhas = sum(linhas)
proportions = [l / total_linhas for l in linhas]

# Colors - using a nice colormap
colors = ['#E63946', "#E06500", '#2A9D8F', "#29A818", '#264653', 
          "#B008DA", '#EE6C4D', '#3D5A80', "#3388B9"]

# Sort by size descending
sorted_indices = np.argsort(linhas)[::-1]
sorted_modulos = [modulos[i] for i in sorted_indices]
sorted_linhas = [linhas[i] for i in sorted_indices]
sorted_arquivos = [arquivos[i] for i in sorted_indices]
sorted_colors = [colors[i] for i in sorted_indices]
sorted_props = [proportions[i] for i in sorted_indices]

def squarify(values, x, y, width, height):
    """
    Squarified treemap algorithm that produces more square-like rectangles.
    Returns list of (x, y, w, h, value, label, color, arqs)
    """
    if not values:
        return []
    
    if len(values) == 1:
        return [(x, y, width, height, values[0]['val'], values[0]['label'], 
                 values[0]['color'], values[0]['arqs'])]
    
    total = sum(v['val'] for v in values)
    
    # Try horizontal split
    if width >= height:
        # Split vertically (left/right)
        # Find split point that makes rectangles most square-like
        best_split = 1
        best_score = float('inf')
        
        for i in range(1, len(values)):
            left_sum = sum(v['val'] for v in values[:i])
            right_sum = sum(v['val'] for v in values[i:])
            
            left_width = width * (left_sum / total)
            right_width = width * (right_sum / total)
            
            # Score based on how close to square
            left_score = abs(left_width - height) / max(left_width, height)
            right_score = abs(right_width - height) / max(right_width, height)
            
            score = left_score + right_score
            
            if score < best_score:
                best_score = score
                best_split = i
        
        left_sum = sum(v['val'] for v in values[:best_split])
        left_width = width * (left_sum / total)
        
        left_rects = squarify(values[:best_split], x, y, left_width, height)
        right_rects = squarify(values[best_split:], x + left_width, y, 
                               width - left_width, height)
        
        return left_rects + right_rects
    
    else:
        # Split horizontally (top/bottom)
        best_split = 1
        best_score = float('inf')
        
        for i in range(1, len(values)):
            top_sum = sum(v['val'] for v in values[:i])
            bottom_sum = sum(v['val'] for v in values[i:])
            
            top_height = height * (top_sum / total)
            bottom_height = height * (bottom_sum / total)
            
            top_score = abs(top_height - width) / max(top_height, width)
            bottom_score = abs(bottom_height - width) / max(bottom_height, width)
            
            score = top_score + bottom_score
            
            if score < best_score:
                best_score = score
                best_split = i
        
        top_sum = sum(v['val'] for v in values[:best_split])
        top_height = height * (top_sum / total)
        
        top_rects = squarify(values[:best_split], x, y + height - top_height, 
                            width, top_height)
        bottom_rects = squarify(values[best_split:], x, y, 
                               width, height - top_height)
        
        return top_rects + bottom_rects

# Prepare data for squarify
data = [{'val': sorted_linhas[i], 'label': sorted_modulos[i], 
         'color': sorted_colors[i], 'arqs': sorted_arquivos[i]} 
        for i in range(len(sorted_modulos))]

# Generate rectangles
rects = squarify(data, 0, 0, 100, 100)

fm.fontManager.addfont('./RobotoSlab.ttf')
fm.fontManager.addfont('./RobotoSlab-Bold.ttf')

matplotlib.rcParams['font.family'] = 'Roboto Slab'

# Create figure
fig, ax = plt.subplots(figsize=(12, 5))

# Draw rectangles
for x, y, w, h, val, label, color, arqs in rects:
    rect = mpatches.FancyBboxPatch(
        (x, y), w, h,
        boxstyle="round,pad=0.02,rounding_size=0.3",
        facecolor=color, edgecolor='white', linewidth=4, alpha=0.9
    )
    ax.add_patch(rect)
    
    # Add text if rectangle is big enough
    min_dim = min(w, h)
    if min_dim > 6:
        fontsize = 16
        
        # Module name (shortened)
        short_name = label
        if short_name == 'lib':
            short_name = 'stepCalc'
        
        ax.text(x + w/2, y + h/2 + 6, short_name, 
               ha='center', va='center', fontsize=fontsize, 
               fontweight="bold", color='white')
        ax.text(x + w/2, y + h/2 - 0, f'{val:,}'.replace(',', '.'), 
               ha='center', va='center', fontsize=fontsize, 
               fontweight='normal',
               color='white')
        # if min_dim > 10:
        ax.text(x + w/2, y + h/2 - 5, f'({arqs} arqs)', 
                   ha='center', va='center', fontsize=fontsize, 
                   color='white')

ax.text(0, 102, "Total: {0:,} linhas".format(total_linhas).replace(',', '.')+f" e {sum(arquivos)} arquivos",
               ha='left', va='bottom', fontsize=20, 
               fontweight="bold")
# ax.text(1.8, 90, "Total: {0:,} linhas".format(total_linhas).replace(',', '.')+f" e {sum(arquivos)} arquivos",
#                ha='left', va='bottom', fontsize=20, 
#                fontweight="bold", bbox=dict(facecolor='white', alpha=0.6, edgecolor='none', pad=10))

# ax.text(0, 102, "Areas proporcionais às linhas de código",
#                ha='left', va='bottom', fontsize=20, 
#                fontweight="bold")

ax.set_xlim(0, 100)
ax.set_ylim(0, 110)
ax.axis('off')

# Title
# Legend
# legend_elements = [mpatches.Patch(facecolor=colors[i], edgecolor='white', alpha=0.9,
                                #   label=f'{modulos[i]} ({linhas[i]:,} linhas, {arquivos[i]} arqs)'.replace(',', '.')) 
                #    for i in range(len(modulos))]
# ax.legend(handles=legend_elements, loc='center', bbox_to_anchor=(0.5, -0.03), 
#           ncol=3, frameon=False, fontsize=9)

plt.tight_layout()
plt.savefig('./dimensao_codigo.png', dpi=300, bbox_inches='tight', 
            facecolor='none', edgecolor='none')
plt.show()
```
